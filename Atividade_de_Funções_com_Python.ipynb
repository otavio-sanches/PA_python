#Função sem Parâmetro e sem Retorno

```python
# Importa o módulo datetime
from datetime import datetime

# Função sem parâmetro e sem retorno
def mostrar_horario():
    agora = datetime.now()  # Pega o horário atual
    print(f"Horário atual: {agora.strftime('%H:%M:%S')}")

# Chamando a função
mostrar_horario()
```

#Função com Parâmetro e sem Retorno

```python
# Função que verifica se um número é par ou ímpar
def verificar_paridade(numero):
    if numero % 2 == 0:
        print(f"O número {numero} é PAR.")
    else:
        print(f"O número {numero} é ÍMPAR.")

# Chamando a função
verificar_paridade(7)
verificar_paridade(12)
```

#Função sem Parâmetro e com Retorno

```python
# Função que soma três notas e retorna o total
def somatresnotas():
    nota1 = float(input("Digite a primeira nota: "))
    nota2 = float(input("Digite a segunda nota: "))
    nota3 = float(input("Digite a terceira nota: "))
    total = nota1 + nota2 + nota3
    return total  # retorna o valor da soma
```

```python

# Função que calcula a média usando o retorno da função anterior
def mediatresnotas():
    soma = somatresnotas()  # usa o retorno da primeira função
    media = soma / 3
    return media  # retorna o resultado da média
```

```python
# Mostrando o resultado final
print("A média das três notas é", mediatresnotas())
```

#Função com Parâmetro e com Retorno

```python
# Função que calcula o total de uma compra
def calcular_total(preco_unitario, quantidade):
    total = preco_unitario * quantidade
    return total

# Programa principal
valor = calcular_total(25.50, 3)
print(f"O valor total da compra é R$ {valor:.2f}")
```

```python
# Função que calcula o IMC
def calcular_imc(peso, altura):
    imc = peso / (altura ** 2)
    return imc

# Programa principal
peso = float(input("Digite seu peso (kg): "))
altura = float(input("Digite sua altura (m): "))

resultado = calcular_imc(peso, altura)
print(f"Seu IMC é {resultado:.2f}")
```

```python
# Função que calcula soma e média de três números
def calcular_soma_e_media(a, b, c):
    soma = a + b + c
    media = soma / 3
    return soma, media  # dois retornos

# Programa principal
n1 = float(input("Digite o primeiro número: "))
n2 = float(input("Digite o segundo número: "))
n3 = float(input("Digite o terceiro número: "))

resultado_soma, resultado_media = calcular_soma_e_media(n1, n2, n3)

print(f"A soma é {resultado_soma:.2f}")
print(f"A média é {resultado_media:.2f}")
```

#2. Conversor de Temperatura

```python
# Conversor de Temperatura
def celsius_para_fahrenheit(c):
    f = c * 9/5 + 32
    return f

# Teste
temp_c = 30
print(f"{temp_c}°C equivalem a {celsius_para_fahrenheit(temp_c)}°F")
```

#4. Quadrado e Cubo

```python
# Quadrado e Cubo
def calcular_potencias(n):
    quadrado = n ** 2
    cubo = n ** 3
    return quadrado, cubo

# Teste
q, c = calcular_potencias(4)
print(f"O quadrado é {q} e o cubo é {c}")
```

#5. Contagem Regressiva

```python
# Contagem Regressiva
def contagem_regressiva(inicio):
    for i in range(inicio, 0, -1):
        print(i)
    print("\nFim da contagem!")

# Teste
contagem_regressiva(5)
```

#6. Cálculo de Desconto

```python
# Cálculo de Desconto
def calcular_desconto(valor_original, percentual_desconto):
    valor_final = valor_original * (1 - percentual_desconto / 100)
    return valor_final

# Teste
print(f"Valor com desconto: R$ {calcular_desconto(100, 15):.2f}")
```

#7. Maior e Menor

```python
# Maior e Menor
def encontrar_extremos(lista):
    maior = max(lista)
    menor = min(lista)
    return maior, menor

# Teste
numeros = [3, 7, 1, 9, 4]
maior, menor = encontrar_extremos(numeros)
print(f"O maior número é {maior} e o menor é {menor}")
```

#8. Calculadora Simples

```python
# Calculadora Simples
def calculadora(a, b, operacao):
    if operacao == "soma":
        return a + b
    elif operacao == "subtracao":
        return a - b
    elif operacao == "multiplicacao":
        return a * b
    elif operacao == "divisao":
        if b != 0:
            return a / b
        else:
            return "Erro: divisão por zero!"
    else:
        return "Operação inválida!"

# Testes
print(calculadora(10, 2, "divisao"))
print(calculadora(10, 2, "multiplicacao"))
```

#10. Sistema de Notas

```python
# Sistema de Notas
def avaliar_aluno(nota):
    if nota >= 7:
        return "Aprovado"
    elif nota >= 5 and nota < 7:
        return "Recuperação"
    else:
        return "Reprovado"

# Testes
print(f"Situação do aluno: {avaliar_aluno(8.5)}")
print(f"Situação do aluno: {avaliar_aluno(6.0)}")
print(f"Situação do aluno: {avaliar_aluno(4.2)}")
```

```python
import random
import string

senha = ""
```

```python
def gerador_senha():
  for i in range(2):
    senha = senha + random.choice(string.ascii_uppercase) + random.choice(string.ascii_lowercase) + random.choice(string.digits) + random.choice(string.punctuacion)
    return(senha)
```

```python
print(gerador_senha)
```

```python
def tabuada(numero):
  for i in range(1,11):
    print(f"{numero} X {i} = {numero * i}")
```

```python
tabuada(6)
```

```python
def validar_senha(senha):
    tamanho_valido = len(senha) >= 8

    tem_numero = False
    tem_maiuscula = False
    tem_minuscula = False
    tem_especial = False

    for char in senha:
        if char.isdigit():
            tem_numero = True

        if char.isupper():
            tem_maiuscula = True

        if char.islower():
            tem_minuscula = True

        if not char.isalnum():
            tem_especial = True

    senha_valida = (
        tamanho_valido and
        tem_numero and
        tem_maiuscula and
        tem_minuscula and
        tem_especial
    )

    return {
        "tamanho_valido": tamanho_valido,
        "tem_numero": tem_numero,
        "tem_maiuscula": tem_maiuscula,
        "tem_minuscula": tem_minuscula,
        "tem_especial": tem_especial,
        "senha_valida": senha_valida
    }
```

```python
validar_senha("0x7dfc24d51440")
```
